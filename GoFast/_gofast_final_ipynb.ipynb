# GoFast_самокаты

## Описание проекта и общий обзор данных

### Описание проекта

__Цель проекта:__ Провести анализ имеющихся данных для сравнения поведения пользователей с разными типами подписки и проверить некоторые гипотезы, которые могут помочь бизнесу вырасти.

__Данные:__ В проекте используются данные о некоторых пользователях популярного сервиса аренды самокатов __GoFast__ из нескольких городов, а также об их поездках в течение 2021 года.

Чтобы совершать поездки по городу, пользователи сервиса GoFast пользуются мобильным приложением. Сервисом можно пользоваться:

- без подписки

абонентская плата отсутствует;

стоимость одной минуты поездки — 8 рублей;

стоимость старта (начала поездки) — 50 рублей;

- с подпиской Ultra

абонентская плата — 199 рублей в месяц;

стоимость одной минуты поездки — 6 рублей;

стоимость старта — бесплатно.

__Задачи:__
1. Найти суммарное расстояние, количество поездок, суммарное время и помесячную выручку для каждого пользователя за каждый месяц;
2. Проверить некоторые гипотезы, которые помогут выяснить важные для дальнейшего развития бизнеса моменты;
3. Визуализировать разницу в поведении пользователей с подпиской и без неё.

### Описание данных

Есть три файла с данными, представляющими информацию о пользователях, их поездках и подписках.

__Пользователи__ — `users_go.csv`

`user_id` -	уникальный идентификатор пользователя

`name` - имя пользователя

`age` -	возраст

`city` - город

`subscription_type` - тип подписки (free, ultra)

_____

__Поездки__ — `rides_go.csv`

`user_id` -	уникальный идентификатор пользователя

`distance` -	расстояние, которое пользователь проехал в текущей сессии (в метрах)

`duration` -	продолжительность сессии (в минутах) — время с того момента, как пользователь нажал кнопку «Начать поездку» до момента, как он нажал кнопку «Завершить поездку»

`date` -	дата совершения поездки

-----

__Подписки__ — `subscriptions_go.csv`

`subscription_type` - тип подписки

`minute_price` - стоимость одной минуты поездки по данной подписке

`start_ride_price` - стоимость начала поездки

`subscription_fee` - стоимость ежемесячного платежа


____

## Предобработка

```python
# приведение столбца date к типу данных datatime
rides['date'] = rides['date'].astype('datetime64[ns]')
rides.info()
```

```python
# создание нового столбца с номером месяца
rides['month'] = rides['date'].dt.month
rides.info()
```

```python
# период поездок
rides['date'].agg(['min', 'max'])
```

```python
rides.head()
```

```python
# обработка явных дубликатов
users = users.drop_duplicates()
users.shape #обнаружились дубликаты
```

```python
rides = rides.drop_duplicates()
rides.shape #дубликатов нет
```

```python
subs = subs.drop_duplicates()
subs.shape #дубликатов нет
```

```python
# округление значения расстояния
rides_rounded = rides.copy()
display(rides_rounded['distance'].head(3))
rides_rounded['distance'] = rides_rounded['distance'].round(2)
rides_rounded['distance'].head(3)
```

```python
# проверка параметра длительности поездок на аномалии
fig = plt.figure(figsize =(5, 8))
plt.boxplot(rides_rounded['duration'])
plt.title("Длительность поездок")
plt.ylabel("длительность, мин")
plt.show()
```

__Итого:__ В процессе предобработки данных выяснилось, что датасет содержит данные о поездках пользователей за 2021 год, а поездки длительностью менее 2 и более 35 минут выглядят аномальными.  Кроме того, были удалены дубликаты в данных.

## Проверка гипотез

Продакт-менеджеры сервиса хотят увеличить количество пользователей с подпиской. Для этого они будут проводить различные акции, но сначала нужно выяснить несколько важных моментов.

6.1 Важно понять, тратят ли пользователи с подпиской больше времени на поездки? Если да, то пользователи с подпиской могут быть «выгоднее» для компании.

6.2 Расстояние одной поездки в 3130 метров — оптимальное с точки зрения износа самоката. Можно ли сказать, что среднее расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров?

6.3 Будет-ли помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки?

6.4 Моделируемая ситуация: техническая команда сервиса обновила сервера, с которыми взаимодействует мобильное приложение. Она надеется, что из-за этого количество обращений в техподдержку значимо снизилось. Некоторый файл содержит для каждого пользователя данные о количестве обращений до обновления и после него. Какой тест понадобился бы для проверки этой гипотезы?

### Проверка гипотезы тратят ли пользователи с подпиской больше времени на поездки?

Необходимо проверить гипотезу о том, тратят ли пользователи с подпиской больше времени на поездки?

Проводим тест для двух выборок: пользователей с подпиской и пользователей без подписки.

В качестве нулевой гипотезы принимаем, что средние значения длительности поездок обеих групп равны, в качестве альтернативной - используем одностороннюю гипотезу «больше», т.е. пользователи с подпиской тратят больше времени на поездки.

```python
# Тратят ли пользователи с подпиской больше времени на поездки?

import scipy.stats as st
#Выбираем уровень статистической значимости 5%
alpha = 0.05

# длительность поездок пользователей с подпиской
ultra_sub_group = data_ultra_sub['duration']
display(f'Среднее значение времени поездок пользователей с подпиской: {data_ultra_sub.duration.mean().round(2)}')

# длительность поездок пользователей без подписки
free_sub_group = data_free_sub['duration']
display(f'Среднее значение времени поездок пользователей без подписки: {data_free_sub.duration.mean().round(2)}')

results = st.ttest_ind(ultra_sub_group, free_sub_group, equal_var=False, alternative='greater')

display(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: пользователи с подпиской тратят больше времени на поездки, чем пользователи без подписки')
else:
    display('Нет оснований отвергнуть нулевую гипотезу')
```

Вывод: статистический тест показывает, что пользователи с подпиской ultra тратят в среднем больше времени на поездки, чем пользователи без подписки.

### Проверка гипотезы что среднее расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров

Необходимо проверить гипотезу о том, что среднее расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров. Расстояние одной поездки в 3130 метров — оптимальное с точки зрения износа самоката.

Проводим тест для одной выборки (пользователи с подпиской). Нулевая гипотеза: среднее расстояние меньше или равно 3130 м, альтернативная - среднее расстояние превышает оптимальное.

```python
# Среднее расстояние с подпиской не превышает 3130 метров

#Выбираем уровень статистической значимости 5%
alpha = 0.05

# расстояние, которое проезжают пользователи с подпиской
ultra_sub_dist = data_ultra_sub['distance']
display(f'Среднее расстояние, которое проезжают пользователи с подпиской: {data_ultra_sub.distance.mean().round(2)}')

# оптимальное расстояние
interested_dist = 3130

results = st.ttest_1samp(ultra_sub_dist, interested_dist, alternative='greater')

display(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: среднее расстояние превышает значение в 3130 м')
else:
    display('Нет оснований отвергнуть нулевую гипотезу: среднее расстояние меньше или равно 3130 м')
```

Вывод: статистический тест показывает, что среднее расстояние, которое проезжают пользователи с подпиской не превышает оптимальное с точки зрения износа самоката значение в 3130 метров.

### Проверка гипотезы что помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки.

Необходимо проверить гипотезу о том, что помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки.

Проводим тест для двух выборок: пользователей с подпиской и пользователей без подписки.

В качестве нулевой гипотезы принимаем, что средние значения помесячной выручки обеих групп равны, в качестве альтернативной - что пользователи с подпиской приносят бОльшую выручку.

```python
# Выручка от пользователей с подпиской выше, чем выручка от пользователей без подписки.
#Выбираем уровень статистической значимости 5%
alpha = 0.05

# помесячная выручка с ultra
ultra_sub_income = data_agg.query('start_ride_price == 0')['income']
display(f'Среднее значение помесячной выручки от пользователей с подпиской: {ultra_sub_income.mean().round(2)}')

# помесячная выручка с free
free_sub_income = data_agg.query('start_ride_price != 0')['income']
display(f'Среднее значение помесячной выручки от пользователей без подписки: {free_sub_income.mean().round(2)}')

results = st.ttest_ind(ultra_sub_income, free_sub_income, equal_var=False, alternative='greater')

display(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: пользователи с подпиской приносят бОльшую выручку.')
else:
    display('Нет оснований отвергнуть нулевую гипотезу: средние значения помесячной выручки обеих групп равны')
```

Вывод: результаты теста показывают, что помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки.

### Проверка гипотезы об уменьшении количества обращений в тех. поддержку

Для проверки гипотезы об уменьшении количества обращений в тех.поддержку применим метод `scipy.stats.ttest_rel()`

В качестве основной гипотезы примем равенство количества обращений до и после обновления, в качестве альтернативной - уменьшение количества обращений после обновления.

## Распределения

### Акция с раздачей промокодов.

Отделу маркетинга GoFast поставили задачу: нужно провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум 100 существующих клиентов должны продлить эту подписку. То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж.

Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают
10 % пользователей.

Необходимо выяснить, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5 %.

Нужно подобрать параметры распределения, описывающего эту ситуацию, построить график распределения и сформулировать ответ на вопрос о количестве промокодов.

```python
from math import sqrt
from scipy import stats as st

n = 1168 # min количество промокодов
p = 0.1 # вероятность продления подписки

# параметры нормального распределения
mu = n * p
sigma = sqrt(n * p * (1 - p))

distr = st.norm(mu, sigma) # функция нормального распределения

result = distr.cdf(100) # вероятность невыполнения плана по продлению подписки (д.б. около 5%)
result
```

```python
from matplotlib import pyplot as plt
from scipy.stats import binom

n = 1168 # количество промокодов
p = 0.1 # вероятность продления подписки

distr = []

# функция биномиального распределения
for k in range(0, n + 1):
    current_value = binom.pmf(k, n, p)
    distr.append(current_value)
# график аппроксимированного нормальными распределения
plt.bar(range(0, n + 1), distr);
```

__ИТОГО:__ 1168 - это минимальное количество промокодов, которые нужно разослать, чтобы с 95% вероятностью получить 100 продлений подписки в рамках акции (или с 5% вероятностью не получить эти 100 продлений).

### Рассылка push-уведомлений

Отдел маркетинга рассылает клиентам push-уведомления в мобильном приложении. Клиенты могут открыть его или не открывать. Известно, что уведомления открывают около 40 % получивших клиентов. Отдел планирует разослать 1 млн уведомлений.

С помощью аппроксимации постройте примерный график распределения и оцените вероятность того, что уведомление откроют не более 399,5 тыс. пользователей.

```python
from math import sqrt
from scipy import stats as st

n = 1000000 # количество рассылаемых уведомлений
p = 0.4 # вероятность открытия уведомления пользователем

# параметры нормального распределения
mu = n * p
sigma = sqrt(n * p * (1 - p))

# функция нормального распределения
distr = st.norm(mu, sigma)

# считаем вероятность хвоста слева от 399,5 тыс
result = distr.cdf(399500)
result
```

__ИТОГО:__ При рассылке пользователям сервиса 1 млн push-уведомлений в мобильном приложении вероятность того, что уведомление откроют не более 399,5 тыс. пользователей равна 15.4%.
