### Описание проекта

Интернет-магазин «Стримчик» продаёт компьютерные игры по всему миру.

Для правильного планирования рекламной кампании на следующий год, данные по продажам игр за предыдущие годы были собраны в файле `games.csv`.

__Цель проекта:__

Выявить определяющие успешность игры закономерности.

Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

__Данные:__

Из открытых источников собраны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

__Задачи:__

В рамках данного проекта необходимо:

- Провести предварительную обработку данных и подготовить их к дальнейшему анализу;

- Проанализировать основные признаки, влияющие на продажи игр;

- Составить портрет пользователя для каждого региона продаж;

- Проверить некоторые гипотезы относительно влияния рейтингов игр на их продажи.

### Описание данных

Набор данных включает следующие поля:
___________
- `Name` — название игры;

- `Platform` — платформа;

- `Year_of_Release` — год выпуска;

- `Genre` — жанр игры;

- `NA_sales` — продажи в Северной Америке (миллионы проданных копий);

- `EU_sales` — продажи в Европе (миллионы проданных копий);

- `JP_sales` — продажи в Японии (миллионы проданных копий);

- `Critic_Score` — оценка критиков (максимум 100);

- `User_Score` — оценка пользователей (максимум 10);

- `Rating` — рейтинг от организации ESRB.

## Предобработка

```python
# перевод названий столбцов в нижний регистр
data.columns = data.columns.str.lower()
data.columns.tolist()
```

### Обработка пропусков и преобразование типов данных

```python
# удаление строк с пропуском года - теряем 1.6%
df = data.dropna(subset=['year_of_release'])
df.shape
```

```python
df['year_of_release'].unique()
```

```python
# преобразование типа данных параметра 'year_of_release'
df['year_of_release'] = df['year_of_release'].astype('int64')
df['year_of_release'].dtypes
```

```python
# преобразование типа данных параметра 'user_score'
df['user_score'].unique() # кроме сомнительного tbd в этом параметре присутствуют и пустые значения
```

```python
# замена tbd на пропуски для дальнейшего изменения типа данных
df['user_score'] = df['user_score'].replace('tbd', np.nan)
df['user_score'].value_counts()
```

```python
df['user_score'] = df['user_score'].astype('float64')
df['user_score'].dtypes
```

```python
#df['user_score'].unique()
```

```python
# описание параметра 'user_score'
df['user_score'].describe()
```

__Итого:__ Необходимость преобразования типов данных в параметрах `year_of_release` и `user_score` привела к обработке пропусков в этих столбцах. Пропуски в `year_of_release` затрудняли последующие исследования динамики продаж по годам, и поскольку логичной замены этим пропускам не нашлось, они были удалены. Потери данных при этом составили 1.6%. После обработки пропусков параметра `year_of_release` тип данных был изменен на int64.

В параметре `user_score` кроме строкового значения tbd были и пропущенные значения. При обработке tbd было заменено пустыми значениями и затем тип данных был заменен на float64. Удаление tdb и пустых значений привело бы к потере более половины данных, а замена нулями или средним значением сильно исказила бы оценки.

```python
df['critic_score'].unique()
```

```python
# описание параметра 'critic_score'
df['critic_score'].describe()
```

```python
# отображение пропущенных значений
display("Missing values:", df.isna().sum())
```

```python
# заполнение пропусков в параметре рейтинга
df['rating'] = df['rating'].fillna('Unknown')
```

```python
# пропуски в жанре
data_na = df[df['genre'].isna()] # без имени и жанра оказалась одна игра - можно удалить
data_na
```

```python
# удаление игры с пропуском в жанре и отображение оставшихся пропусков
df = df.dropna(subset=['genre'])
display("Missing values:", df.isna().sum())
```

```python
# отображение пропусков после обработки
sns.heatmap(df.isna(), yticklabels=False, cbar=False, cmap=colors);
```

__Итого:__ Пропуски в параметре `critic_score` остались без изменения, поскольку логической замены им не нашлось, а замена какими-либо значениями повлияет на дальнейшие исследования.

Пропущенные значения в параметре `rating` были заменены на 'Unknown'.

Две игры с пропусками в названиях и жанре были удалены. Как оказалось, у этих игр отсутствовали так-же рейтинг и оценки.

Вероятно, большое количество пропусков в данных вызвано сбором данных из разных источников, где разные наборы данных могли храниться в разных форматах.

- Анализ аномалий показал, что в параметрах продаж (особенно суммарных) выделяются экстремальные значения, похожие на выбросы. Реальность этих цифр (продажи самых коммерчески успешных игр) была подтверждена выборочной проверкой, поэтому экстремальные значения остались в датафрейме. Однако, лишь 5% игр были проданы в объеме чуть больше 2 млн копий и только 1% - 5.5 млн копий и больше.

- Матрица корреляции числовых параметров итогового датафрейма показывает значительную корреляцию параметров продаж (по регионам и общих), а также заметную корреляцию оценок критиков и пользователей. Кроме того, заметна некоторая обратная корреляция оценок пользователей и года выпуска игры, т.е. более поздние игры имеют более низкие оценки. Возможно, пользователи становятся требовательнее.

## Проверка гипотез

### Проверка гипотезы о средних пользовательских оценках игр, выпущенных для платформ Xbox One и PC

Необходимо проверить гипотезу, что средние пользовательские оценки игр, выпущенных для платформ Xbox One и PC одинаковые.

Проводим тест для двух выборок: пользователей с приставками XBox, и пользователей с РС.

В качестве нулевой гипотезы принимаем, что средние пользовательские оценки игр, выпущенных для платформ Xbox One и PC равны, в качестве альтернативной - используем двустороннюю гипотезу, т.е. средние пользовательские оценки этих игр не равны.

```python
# проверка гипотезы о равенстве пользовательских оценок платформ
import scipy.stats as st
# выбираем уровень статистической значимости 5%
alpha = 0.05
```

```python
actual_df['platform'].unique()
```

```python
# датафрейм, где пользовательские оценки определены
scored_df = actual_df[actual_df['user_score'].notna()]
#scored_df
```

```python
# выборка XOne
xbox_data = scored_df.query('platform == "XOne"')
#xbox_data
```

```python
# выборка PC
pc_data = scored_df.query('platform == "PC"')
#pc_data
```

```python
# средние пользовательские оценки платформы Xbox
xbox_value = xbox_data['user_score']
display(f'Среднее значение пользовательского рейтинга Xbox One: {xbox_data.user_score.mean().round(2)}')

# средние пользовательские оценки платформы PC
pc_value = pc_data['user_score']
display(f'Среднее значение пользовательского рейтинга PC: {pc_data.user_score.mean().round(2)}')

results = st.ttest_ind(xbox_value, pc_value, equal_var=False)

display(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: средние пользовательские рейтинги обеих платформ не одинаковые')
else:
    display('Нет оснований отвергнуть нулевую гипотезу')
```

__Итого:__ Сравнение средних значений пользовательского рейтинга игр для платформ Xbox One и РС показывает, что средний рейтинг игр для Xbox One чуть выше, чем для РС.

Однако, использование статистического теста для проверки гипотезы о равенстве этих значений показывает, что нет оснований отвергнуть нулевую гипотезу о равенстве значений. Таким образом, подтверждается гипотеза, что средние пользовательские оценки игр, выпущенных для платформ Xbox One и PC одинаковые.

### Проверка гипотезы о средних пользовательских оценках игр в жанрах Action и Sports

Необходимо проверить гипотезу, что средние пользовательские оценки игр в жанрах Action и Sports не одинаковые.

Проводим тест для двух выборок: игры в жанре Action и игры в жанре Sports.

В качестве нулевой гипотезы принимаем, что средние пользовательские оценки игр в жанрах Action и Sports равны, в качестве альтернативной - используем двустороннюю гипотезу, т.е. средние пользовательские оценки игр в этих жанрах не равны.

Таким образом, если будут основания отвергнуть нулевую гипотезу, неравенство средних оценок будет подтверждено.

```python
# проверка гипотезы о неравенстве пользовательских оценок жанров
# выбираем уровень статистической значимости 5%
alpha = 0.05
```

```python
scored_df['genre'].unique()
```

```python
action_data = scored_df.query('genre == "Action"')
#action_data
```

```python
sports_data = scored_df.query('genre == "Sports"')
#sports_data
```

```python
# средние пользовательские оценки игр в жанре Action
action_value = action_data['user_score']
display(f'Среднее значение пользовательского рейтинга жанра Action: {action_data.user_score.mean().round(2)}')

# средние пользовательские оценки платформы PC
sports_value = sports_data['user_score']
display(f'Среднее значение пользовательского рейтинга жанра Sports: {sports_data.user_score.mean().round(2)}')

results = st.ttest_ind(action_value, sports_value, equal_var=False)

display(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: средние пользовательские рейтинги обоих жанров не одинаковые')
else:
    display('Нет оснований отвергнуть нулевую гипотезу')
```

__Итого:__ Сравнение средних значений пользовательских оценок игр в жанре Action и Sports показывает, что средние оценки игр в жанре Action выше, чем  средние оценки игр в жанре Sports.

Использование статистического теста для проверки гипотезы о равенстве этих значений подтверждает, что нулевая гипотеза о равенстве значений отвергается.

Таким образом, можно заключить, что средние пользовательские оценки игр в жанрах Action и Sports не равны.
